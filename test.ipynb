```python
import json
from src.ingestor.ingerir_texto import IngestorTexto
from src.ingestor.ingerir_imagem import IngestorImagem
from src.ingestor.ingerir_tabela import IngestorTabela
from src.vector_db.chroma_client import ChromaDB
from src.embeddings.modelo_texto import ModeloEmbeddingTexto
from src.embeddings.modelo_imagem import ModeloEmbeddingImagem

ingestor_texto = IngestorTexto()
ingestor_imagem = IngestorImagem()
ingestor_tabela = IngestorTabela()
modelo = ModeloEmbeddingTexto()
modelo_imagem = ModeloEmbeddingImagem()
chroma = ChromaDB()

arquivo = "data/aspectos_do_planejamento_orcamento_dos_estados_rio_grande_do_sul.json"
```

```python
documentos_textos = ingestor_texto.processar_json(arquivo)
documentos_imagens = ingestor_imagem.processar_json(arquivo)
documentos_tabelas = ingestor_tabela.processar_json(arquivo)
```

```python
for documento in documentos_textos:
    chroma.inserir_documento(documento)

for documento in documentos_imagens:
    chroma.inserir_imagem(documento)

for documento in documentos_tabelas:
    chroma.inserir_tabela(documento)
```

```python
pergunta = "Quuais são as prioridades do projeto Rio Grande durante 1971 e 1974?"
retrieved = chroma.busca_completa(pergunta, modelo, modelo_imagem, 5)
```
